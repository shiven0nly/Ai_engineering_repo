#### 2D Line plot
* use for Bivariate Analysis
* Both cols numerical, one col is numerical and other categorical
* Use case -> Time series data (like time ke basis per analysis krte hai)

```python
# plotting a simple function
price=[100,130,140,150,145,137]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price) # (x,y)
```

```python
# from a pandas dataframe
sharma_kohli=pd.read_csv('./datasets_1/sharma-kohli.csv')
sharma_kohli
```

```python
plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'])
```

plotting multiple plots , creating vkohli and rgsharma plot simultaneaously.

```python
plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'])
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'])
```

labels titles

```python
plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'])
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'])

plt.title("Rohit Sharma vs Virat Kohli Comparison")
plt.xlabel('Season')
plt.ylabel('runs scored')
```

we can also change the color, we can also provide hex code

```python
plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'],color='green')
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'],color='red')

plt.title("Rohit Sharma vs Virat Kohli Comparison")
plt.xlabel('Season')
plt.ylabel('runs scored')
```

we can also change the solid line of graph to dotted , dashed & dashdot

```python
plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'], linestyle='dotted')
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'],linestyle='dashed')

plt.title("Rohit Sharma vs Virat Kohli Comparison")
plt.xlabel('Season')
plt.ylabel('runs scored')
```

we can also manipulate the width of the lines

```python
plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'],linewidth='2')
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'],linewidth='5')

plt.title("Rohit Sharma vs Virat Kohli Comparison")
plt.xlabel('Season')
plt.ylabel('runs scored')
```

we can also make the points highlight usking marker

```python
plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'],marker='D') # D = diamond
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'],marker='o') # o = circle

plt.title("Rohit Sharma vs Virat Kohli Comparison")
plt.xlabel('Season')
plt.ylabel('runs scored')
```

how to label the legends : like which graph shows which thing? so we use plt.legend() for this

```python
plt.plot(sharma_kohli['index'],sharma_kohli['V Kohli'],label='Virat')
plt.plot(sharma_kohli['index'],sharma_kohli['RG Sharma'],label='Rohit')

plt.title("Rohit Sharma vs Virat Kohli Comparison")
plt.xlabel('Season')
plt.ylabel('runs scored')
plt.legend()
```

### Bar Chart
* Bivariate analysis
* Numerical vs Categorical
* Use Case = Aggregrate analysis of groups

```python
# simple bar chart
plt.style.use('bmh')
children=[10,20,12,15]
colors=['red','blue','yellow','pink']
plt.bar(colors,children)
plt.show()
```

```python
# for horizontal bar chart
plt.barh(colors,children,color='green')
```

multiple bar charts / side by side bar graphs

```python
bats=pd.read_csv('./datasets_1/batsman_season_record.csv')
bats
```

```python
plt.bar(np.arange(bats.shape[0])-0.2,bats['2015'],width=0.2,color='violet')
plt.bar(np.arange(bats.shape[0]),bats['2016'],width=0.2,color='blue')
plt.bar(np.arange(bats.shape[0])+0.2,bats['2017'],width=0.2,color='green')
plt.xticks(np.arange(bats.shape[0]),bats['batsman'])
plt.show()
```

stacked bar chart, using bottom parameter

```python
plt.bar(bats['batsman'],bats['2017'], label='2017')
plt.bar(bats['batsman'],bats['2016'],bottom=bats['2017'],label='2018')
plt.bar(bats['batsman'],bats['2015'],bottom=(bats['2016']+bats['2017']),label='2015')
plt.legend()
plt.show()
```

### Pie Chart
* Univarite and Bivariate
* Categorical and numerical
* Use case: To find contribution on a scale of 100

```python
gayle=pd.read_csv('./datasets_1/gayle-175.csv')
gayle
```

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'])
# for percentage contribution
# autopct='%0.1f%%'
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# we can show one section by taking out it using explode[0,1,0,1], 1 => explode TRUE, 0=> explode FALSE
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],explode=[1,0,0,0,0,0])
```
