find the highes rated movie of each genre

```python
# we can use this using a loop
df = pd.DataFrame(columns=movies.columns)
for group,data in genres:
    df = pd.concat(data[data['IMDB_Rating']==data['IMDB_Rating'].max()])
df
```

split -> (apply) -> combine
apply -> builtin function

```python
movies.groupby('Genre').apply(min)
```

find ranking of each movie in the group according to IMDB score

```python
def rank_movie(group):
    group['genre_rank']=group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genres.apply(rank_movie)
```

find normalized IMDB rating group wise

```python
def normalized_rating(group):
    group['Normalized_rating']=(group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(normalized_rating)
```

find the best (in-terms of metascore(avg)) actor-> genre combo

```python
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False)
```

### IPL DATASET

```python
ipl = pd.read_csv('./datasets_3/deliveries.csv')
ipl.head()
```

find the top 10 batsman in terms of run

```python
# 1. we divide the whole data in terms of groups according to batsman
# 2. hr batsman jitni balls kheli un sbko alg kr rhe hai
# 3. and batsman_run wale column ko lenge and sum() kr denge
# 4. and it will give us the total run made by that batsman

ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

find the batsman with max no of sixes

```python
six=ipl[ipl['batsman_runs']==6]
```

```python
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

find batsman with most number of 4's and 6's in last 5 overs

```python
last_five=ipl[ipl['over']>15]
last_five[(last_five['batsman_runs'] == 4) | (last_five['batsman_runs'] == 6)]
last_five.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(5)
```
