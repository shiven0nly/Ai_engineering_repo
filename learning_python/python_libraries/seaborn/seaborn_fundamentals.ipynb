```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
tips.head(5)
```

scatter plot

```python
# x = bill
# y = tip
sns.scatterplot(data=tips, x="total_bill", y="tip")
# yeh ek 'axes level' plot hai
```

```python
# replot -> figure level -> square shape hoti hai

sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter')
```

we want the information that dot is male or female
-> we use hue parameter that gives the different values with different colors

```python
sns.scatterplot(data=tips, x="total_bill", y="tip",hue='sex')
```

```python
# with time we can see that different customers are represented with different types
sns.scatterplot(data=tips, x="total_bill", y="tip",hue='sex',style='time')
```

#### Line plot

```python
gap=px.data.gapminder()
gap
```

Above data se hume India ki lifeExpectancy use krna hai

```python
df=gap[gap['country']=='India']
sns.lineplot(data=df,x='year',y='lifeExp')
```

facet plot: like we have to make different graphs for different variables for example in the tips data, we have to make seperate graph for male and female , so what we do is ? we use col='sex' parameter, and it make two graphs parallely to each other, if we pass row='sex' then it will make two graphs one up and one down

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex')
```

### 2. Distribution plot
* used for univariate analysis
* used to find out the distribution
* range of the observation
* central tendancy
* is the data bimodal? (multiple peaks in graphs)
* are there outliers?(points that lie far away from the rest of the numbers in a set)

```python
# figure level -> displot
# axes level -> name_of_plot
```

```python
# plotting univariate hist
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='hist')
```

kde= kernal destiny estimation, the observations with a Gaussian curve.

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',fill=True)
```

Bivariate histogram: a bivariate hist bins the data within rectangles that title the plot
and then shows the count of observations within each rectangle with the fill color

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('bmh')
tips=sns.load_dataset('tips')
sns.displot(data=tips,x='total_bill',y='tip',kind='hist')
```

Bivariate kdeplot
-> a bivariate kde plot smoothes the (x,y) observations with a 2D guassina

```python
sns.displot(data=tips,x='total_bill',y='tip',kind='kde')
```

#### Matrix
* Heatmap
* Clustermap

```python
# Heatmap
df=gap.pivot(index='country',columns='year',values='lifeExp')

# axes level function hota hai , isme koi figure level function nhi hota
sns.heatmap(df)
```

we can change the axes level graph size using the function:
plt.figure(figsize=(x,y))

```python
plt.figure(figsize=(15,20))
sns.heatmap(df)
plt.show()
```

# annot, for accurate results

```python
df=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(df,annot=True,linewidths=0.5,cmap='autumn')
```
