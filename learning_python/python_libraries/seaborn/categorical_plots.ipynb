# In this we cover the categorical plots, regplot and multiplots

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

strip plot: yeh ek numerical col ko leta hai and ek categorical col ko leta hai scatter plot bna deta hai

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',hue='day')
# we using catplot for figure level functionn
```

swarmplot: algo run, that it represents the distribution

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm',hue='day')
```

##### BOXPLOT
-> a boxplot is a standardized way of displaying the distribution of data based on a five number summary ('minimum', first quartile[Q1], median, third quartile[Q3] and maximum). it can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

```python
from IPython.display import Image
image_url='https://www.simplypsychology.org/wp-content/uploads/image-3.jpeg'
Image(url=image_url, width=500)
```

Q1 = 25th percentile
Q3 = 75th percentile
IQR = interquartile rang
min = Q1 - 1.5*IQR (not literal min, but calculated min, possible values are greater or lesser than this min.)
max = Q3 + 1.5*IQR

```python
sns.set_theme(style='whitegrid',palette='pastel')
sns.catplot(data=tips,x='sex',y='total_bill',hue='sex',kind='box')
```

```python
sns.catplot(data=tips,y='total_bill',kind='box')
```

Violin plot: boxplot + kde plot

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='day')
```

Barplot: we can use aggregate methods on it

```python
import numpy as np
sns.catplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.mean,kind='bar')
```
