#### Types of Unclean Data
There are 2 kinds of unclean data
* Dirty data (Data with Quality issues): Dirty dat, also known as low quality data.Low quality data has content issues
- Duplicated data
- Missing data
- Corrupt data
- Ineqaulity data

* Messy data(Data with tidiness issues): Messy data, also known as untidy data. untidy data has structural issues. Tidy data has the following properties:
- Each variable forms a column
- Each observation forms a row
- Each observational unit forms a table

#### Data Quality Dimensions:-
* Completeness -> is data missing?
* Validity -> is data invalid -> negative height -> duplicate patient id
* Accuracy -> data is valid but not accurate -> weight-> 1kg
* Consistency -> both valid accurate but written differently -> New York and NY

### Order of Severity:
completeness > Validity > Accuracy > Consistency

#### ORDER:
1. Completeness
2. Tidy
3. Validity
4. Accuracy
5. Consistency

#### Tidiness: Messy data

Define:
* patients table me contact col me phone and email dono sath me hai
* we have to seperate them out using regex email and phone

```python
# its out of our scope for now
```

```python
# Auralin and Novodra must be in different cols for treatment, treatment_cut
treatment=pd.concat([treatments,treatments_cuts])
treatment
```

```python
treatment=treatment.melt(id_vars=['given_name',	'surname', 'hba1c_start',	'hba1c_end',	'hba1c_change'],var_name='dose_type',value_name='doses')
```

```python
treatment=treatment[treatment['doses']!='-']
treatment
```

```python
# doses are starting and ending values we have to seperate them out

treatment['dose_start']=treatment['doses'].str.split('-').str.get(0)
treatment['dose_end']=treatment['doses'].str.split('-').str.get(1)
treatment
```

```python
# we can also remove the u
treatment['dose_start']=treatment['dose_start'].str.replace('u','')
treatment['dose_end']=treatment['dose_end'].str.replace('u','')
treatment['dose_start']=treatment['dose_start'].astype('int')
treatment['dose_end']=treatment['dose_end'].astype('int')
treatment.info()
```

Adverse table shouldnt exist independently as its only taking more space

```python
treatment=pd.merge(treatment,adverse)
treatment
```
